```python
inList = data["inList"]
inWeight = data["inWeight"]
outList = data["outList"]
outWeight = data["outWeight"]
usernameList = data["usernameList"]
```

```python
combined = collections.Counter(usernameList + mapped_members)
filtered_counter = collections.Counter({user: count for user, count in combined.items() if count > 1})
```

```python
caucus_color_mapping = {"Progressive Caucus": "blue",
                       "New Democratic Coalition": "lightblue",
                       "Problem Solvers Caucus": "purple",
                       "Republican Governance Group":"lightred",
                       "Republican Study Committee":"red",
                       "Freedom Caucus":"darkred",
                       "None":"grey"}
```

```python
G = nx.Graph()

for user in usernameList:
  caucus = member_caucus_mapping.get(user)
  G.add_node(user, caucus=caucus)
```

```python
for i, targets in enumerate(outList):
    user = usernameList[i]
    for target_idx in targets:
        target = usernameList[target_idx]
        if user in G.nodes and target in G.nodes:
            G.add_edge(user, target)
```

```python
node_colors = [caucus_color_mapping[member_caucus_mapping[node]] for node in G.nodes]

plt.figure(figsize=(20, 10))
nx.draw(G, with_labels=False, node_color=node_colors, edge_color="gray", node_size=150, alpha=0.5, font_color = "black")
plt.show()
```
